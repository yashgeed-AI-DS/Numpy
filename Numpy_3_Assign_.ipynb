# NumPy Assignment — Array Operations, Broadcasting & Aggregate Functions




# Q1. Array Addition

## Problem Statement
Create two NumPy arrays:

```python
arr1 = [1, 2, 3, 4]
arr2 = [10, 20, 30, 40]
```

Add both arrays element-wise.

## Sample Output
```python
[11 22 33 44]
```


````python
import numpy as np
arr1 = np.array([1,2,3,4])
arr2 = np.array([10,20,30,40])
print(arr1+arr2)
````


# Q9. Row-wise Sum using Axis

## Problem Statement
Create the array:

```python
[[1,2,3],
 [4,5,6]]
```

Find row-wise sum using axis.

## Sample Output
```python
[ 6 15]
```


````python
a = np.array([[1,2,3],
              [4,5,6]])
print(np.sum(a,axis = 1))
````
